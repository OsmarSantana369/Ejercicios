```python
# Vamos a cargar base2.xlsx y hacer un análisis exploratorio de datos (EDA)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Cargar el archivo Excel
data = pd.read_excel('base2.xlsx')
# Transponer el DataFrame para facilitar el análisis
data = data.transpose()
# Mostrar las primeras filas del DataFrame
print(data.head())
```

```python
# Eliminar la primera, segunda, y cuarta filas
data = data.drop(index=['Country Name', 'Country Code', 'Series Code'], errors='ignore')
data.reset_index(drop=True, inplace=True)   # Reiniciar el índice después de eliminar filas
# Poner la primera fila como encabezado
data.columns = data.iloc[0]
data = data[1:]  # Eliminar la fila que ahora es el encabezado
print(data.head())
```

```python
# Por cada columna, mostrar las medidas de tendencia central y dispersión, asi como una representación gráfica, omitiendo los valores nulos
for column in data.columns:
	print(f"Análisis de la variable: {column}")
	print(f"Media: {data[column].mean()}")
	print(f"Mediana: {data[column].median()}")
	print(f"Moda: {data[column].mode()[0]}")
	print(f"Desviación estándar: {data[column].std()}")
	print(f"Varianza: {data[column].var()}")
	print(f"Mínimo: {data[column].min()}")
	print(f"Máximo: {data[column].max()}")
	print(f"Rango: {data[column].max() - data[column].min()}")
	print("\n")
	
	# Cada fila representa un año (las etiquetas deben empezar desde 2000), por lo que haremos un gráfico de barras (usando sns) para cada variable
	plt.figure(figsize=(10, 6))
	sns.barplot(x=range(2000, 2000 + len(data.index)), y=data[column])
	plt.title(f"{column} a lo largo de los años")
	plt.xlabel("Año")
	plt.ylabel("")
	plt.xticks(rotation=45)
	plt.show()
```

```python
# Se eliminarán las primeras cuatro columnas y la número 13 hacia la 17
data_bivariado = data.drop(columns=data.columns[:4].tolist() + data.columns[12:17].tolist(), errors='ignore')
mujeres = data_bivariado[data_bivariado.columns[[0, 2, 4, 6]]]
hombres = data_bivariado[data_bivariado.columns[[1, 3, 5, 7]]]
```

```python
# Obtener el coeficiente de correlación de Pearson entre la variable de la primer columna del Data Frame de mujeres y el resto de columnas
for column in mujeres.columns[1:]:
	correlation = mujeres[mujeres.columns[0]].astype(float).corr(mujeres[column].astype(float))
	print(f"Coeficiente de correlación de Pearson entre {mujeres.columns[0]} y {column} es: {correlation}")

# Obtener el coeficiente de correlación de Pearson entre la variable de la primer columna del Data Frame de hombres y el resto de columnas
for column in hombres.columns[1:]:
	correlation = hombres[hombres.columns[0]].astype(float).corr(hombres[column].astype(float))
	print(f"Coeficiente de correlación de Pearson entre {hombres.columns[0]} y {column} es: {correlation}")
```

Como se puede observar, todos los coeficientes de correlación son altos, lo que indica que hay una relación lineal fuerte entre la variable de desempleo y las demás variables relacionadas a la educación, tanto en mujeres como en hombres.

A continuación, se presentan los gráficos de dispersión para cada par de variables analizadas junto con la línea de regresión lineal ajustada:

```python
# Generar gráficos de dispersión con línea de regresión para cada par de variables analizadas
for column in mujeres.columns[1:]:
	plt.figure(figsize=(8, 6))
	sns.regplot(x=mujeres[mujeres.columns[0]].astype(float), y=mujeres[column].astype(float))
	plt.title(f"Gráfico de dispersión entre {mujeres.columns[0]} y \n {column}")
	plt.xlabel("")
	plt.ylabel("")
	plt.show()

for column in hombres.columns[1:]:
	plt.figure(figsize=(8, 6))
	sns.regplot(x=hombres[hombres.columns[0]].astype(float), y=hombres[column].astype(float))
	plt.title(f"Gráfico de dispersión entre {hombres.columns[0]} y \n {column}")
	plt.xlabel("")
	plt.ylabel("")
	plt.show()
```
