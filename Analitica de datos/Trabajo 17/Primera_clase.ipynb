```python
#Generalemnte vamos a trabajar con vectores
vector1=np.array([1,2,3,4,5])
print(vector1)
vector2=np.array([6,7,8,9,10])
print(vector2)
```

```python
#Mostrar las operaciones básicas con estos dos vectores
print("La suma es:", vector1+vector2)
print("La resta es:",vector1-vector2)
print("La multiplicación entrada a entrada:", vector1*vector2)
print("La división entrada a entrada:",vector1/vector2)
```

```python
#Producto escalar y vectorial de los dos vectores que tenemos como ejemplo.
print("El producto escalar es:",np.dot(vector1,vector2))
#print("El producto vectorial es:",np.cross(vector1,vector2)) #no es valido fuera de R^3
```
